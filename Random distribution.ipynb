```python
import numpy as np
np.random.seed(123)
all_walks = []
# Simulate random walk 10 times
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
            if np.random.rand() <= 0.001 :        #resetting the step to Zero there is clumpsy due to values b/w 0 to 1
                 step = 0
        random_walk.append(step)
    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
#print(all_walks)
import matplotlib.pyplot as plt
plt.hist(all_walks,50)
plt.show()
plt.clf()
np_allw = np.array(all_walks)
np_allt = np.transpose(np_allw)

plt.plot(np_allt)
plt.show()
```
